# Project Formulation: Provisional Lane Selection

1. Selected Lane

* Lane Choice: CTR / Engagement Opportunity Scoring

* Type: Core Lane

2. Problem Framing

* Search Question: Which high-volume pages or search queries are underperforming in click-through rate (CTR) compared to their search position, and how can we rank them to fix them first?
* Unit of Analysis: A page-query pair.
* Output: A volume-adjusted opportunity score that ranks pages from highest to lowest potential traffic lift.
* Actionable Impact: Once the model spits out the top 50 underperforming pages each week, the SEO and content teams can immediately step in to rewrite messy titles, fix meta descriptions, or tweak the page layouts to claw back that missing traffic.
* Cost of a Wrong Recommendation: If the model gets it wrong, we waste valuable engineering and design hours fixing pages that didn't need any work in the first place—or worse, we accidentally mess with the layout of a high-performing page and tank its search ranking.
* Why Data/ML Helps: Click-through rates drop off a cliff from position 1 to position 3, so it’s not a straight line. Because search positions and user behavior change constantly, we can't just use a basic rule or a hardcoded formula to catch these issues. A machine learning model can actually handle these uneven patterns and help us predict what a page should be scoring versus what it's actually getting.
